```python

import hashlib
import os

def calculate_hash(filepath):
    hasher = hashlib.sha256()
    with open(filepath, 'rb') as file:
        while chunk := file.read(8192):
            hasher.update(chunk)
    return hasher.hexdigest()

def monitor_files(file_list):
    hashes = {}
    for file in file_list:
        if os.path.exists(file):
            hashes[file] = calculate_hash(file)
        else:
            print(f"File not found: {file}")
    return hashes

def detect_changes(old_hashes):
    changes = []
    for file, old_hash in old_hashes.items():
        if os.path.exists(file):
            new_hash = calculate_hash(file)
            if new_hash != old_hash:
                changes.append(file)
        else:
            changes.append(f"{file} (Deleted)")
    return changes

# Example usage
files = ['example.txt']
initial_hashes = monitor_files(files)

input("Make changes to the files, then press Enter...")

changed_files = detect_changes(initial_hashes)
print("Changed files:", changed_files)
```
